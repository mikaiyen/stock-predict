```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('df2330_13to22.csv',thousands=',')
data
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data['自營商買進tmpsum']=data['自營商買進股數(自行買賣)'].fillna(0)+data['自營商買進股數(避險)'].fillna(0)
data['自營商賣出tmpsum']=data['自營商賣出股數(自行買賣)'].fillna(0)+data['自營商賣出股數(避險)'].fillna(0)
```

```python
data=data.drop(['證券代號','證券名稱','自營商買進股數(自行買賣)','自營商買進股數(避險)','自營商賣出股數(自行買賣)','自營商賣出股數(避險)','自營商買賣超股數(自行買賣)','自營商買賣超股數(避險)'],axis=1)
```

```python
data['自營商買進股數']=data['自營商買進股數'].fillna(data['自營商買進tmpsum'])
data['自營商賣出股數']=data['自營商賣出股數'].fillna(data['自營商賣出tmpsum'])
```

```python
data
```

```python
data=data.drop(['自營商買進tmpsum','自營商賣出tmpsum'],axis=1)
```

```python
data.isnull().sum()
```

```python
data['外資買進股數']=data['外資買進股數'].fillna(data['外陸資買進股數(不含外資自營商)'])
data['外資賣出股數']=data['外資賣出股數'].fillna(data['外陸資賣出股數(不含外資自營商)'])
data['外資買賣超股數']=data['外資買賣超股數'].fillna(data['外陸資買賣超股數(不含外資自營商)'])
```

```python
data=data.drop(['外陸資買進股數(不含外資自營商)','外陸資賣出股數(不含外資自營商)','外陸資買賣超股數(不含外資自營商)'],axis=1)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data['DateTime']=pd.to_datetime(data['Date'], format='%Y%m%d')
data
```

```python
cols=data.columns.tolist()
cols=cols[-1:]+cols[:-1]
data=data[cols]
```

```python
data.rename(columns={'成交股數': 'Volume', '成交筆數': 'Transaction', '成交金額': 'Value', '開盤價': 'Open', '最高價': 'High', '最低價': 'Low', '收盤價': 'Close', '漲跌(+/-)': 'Dir', '漲跌價差': 'Change', '外資買進股數': 'ForeignBuy', '外資賣出股數': 'ForeignSell', '外資買賣超股數': 'ForeignNet', '投信買進股數': 'TrustBuy', '投信賣出股數': 'TrustSell', '投信買賣超股數': 'TrustNet', '自營商買賣超股數': 'DealerNet', '自營商買進股數': 'DealerBuy', '自營商賣出股數': 'DealerSell', '三大法人買賣超股數': 'InvestorsNet', '本益比': 'PE-Ratio', '股價淨值比': 'PBR', '殖利率': 'Yield'}, inplace=True)
data
```

```python
# cols=data.columns.tolist()
# cols=cols[:8]+cols[9:]+cols[8:9]
# data=data[cols]
data['ForeignVol']=data['ForeignBuy']+data['ForeignSell']
data['TrustVol']=data['TrustBuy']+data['TrustSell']
data['DealerVol']=data['DealerBuy']+data['DealerSell']
data['InvestorsVol']=data['ForeignVol']+data['TrustVol']+data['DealerVol']
data
```
