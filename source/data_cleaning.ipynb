```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('dfnew_proc_v3.csv',thousands=',')
print(data.head())
```

```python
data.info()
```

```python
data['Volume']=(data['Volume']/1000).astype(int)
data['ForeignBuy']=(data['ForeignBuy']/1000).astype(int)
data['ForeignSell']=(data['ForeignSell']/1000).astype(int)
data['ForeignNet']=(data['ForeignNet']/1000).astype(int)
data['TrustBuy']=(data['TrustBuy']/1000).astype(int)
data['TrustSell']=(data['TrustSell']/1000).astype(int)
data['TrustNet']=(data['TrustNet']/1000).astype(int)
data['DealerNet']=(data['DealerNet']/1000).astype(int)
data['DealerBuy']=(data['DealerBuy']/1000).astype(int)
data['DealerSell']=(data['DealerSell']/1000).astype(int)
data['InvestorsNet']=(data['InvestorsNet']/1000).astype(int)

data=data.drop(['Value'],axis=1)
data
```

```python
# trend 0:漲跌在5%內/ 1:漲超過5% /2:跌超過5%

# data['trend'] = np.where(stock_data.Close.shift(-5) > stock_data.Close, 1, 0)

data.loc[data.Close.shift(-5) > data.Close, "trend"]= 1
data.loc[data.Close.shift(-5) == data.Close, "trend"]= 0
data.loc[data.Close.shift(-5) < data.Close, "trend"]= 2
data.loc[((data.Close.shift(-5)/data.Close)<=1.03)&((data.Close.shift(-4)/data.Close)>=0.97), "trend"]= 0

data['trend']=data['trend']
data['trend'] = data['trend'].fillna(0)
print(data['trend'].value_counts())
data
```

```python

# data.loc[data["InvestorsNet"] == 0, "InvestorsNet_Dir"]= 0
data.loc[data["InvestorsNet"] >= 0, "InvestorsNet_Dir"]= 1
data.loc[data["InvestorsNet"] < 0, "InvestorsNet_Dir"]= 0
data['InvestorsNet_Dir']=data['InvestorsNet_Dir'].astype(int)

# data.loc[data["ForeignNet"] == 0, "ForeignNet_Dir"]= 0
data.loc[data["ForeignNet"] >= 0, "ForeignNet_Dir"]= 1
data.loc[data["ForeignNet"] < 0, "ForeignNet_Dir"]= 0
data['ForeignNet_Dir']=data['ForeignNet_Dir'].astype(int)

# data.loc[data["TrustNet"] == 0, "TrustNet_Dir"]= 0
data.loc[data["TrustNet"] >= 0, "TrustNet_Dir"]= 1
data.loc[data["TrustNet"] < 0, "TrustNet_Dir"]= 0
data['TrustNet_Dir']=data['TrustNet_Dir'].astype(int)

# data.loc[data["DealerNet"] == 0, "DealerNet_Dir"]= 0
data.loc[data["DealerNet"] >= 0, "DealerNet_Dir"]= 1
data.loc[data["DealerNet"] < 0, "DealerNet_Dir"]= 0
data['DealerNet_Dir']=data['DealerNet_Dir'].astype(int)

# data=data.drop(['InvestorsNet','ForeignNet','TrustNet','DealerNet'],axis=1)
data
```

```python
data['Dir'].value_counts()
```

```python

# Dir2 0:無漲跌/ 1:漲 /2:跌

data.loc[data.Close.shift(1) > data.Close, "Dir2"]= 2
data.loc[data.Close.shift(1) == data.Close, "Dir2"]= 0
data.loc[data.Close.shift(1) < data.Close, "Dir2"]= 1
data['Dir2'] = data['Dir2']

data['Dir2'] = data['Dir2'].fillna(0)
data
```
