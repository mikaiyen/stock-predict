```python
data['自營商買進tmpsum']=data['自營商買進股數(自行買賣)'].fillna(0)+data['自營商買進股數(避險)'].fillna(0)
data['自營商賣出tmpsum']=data['自營商賣出股數(自行買賣)'].fillna(0)+data['自營商賣出股數(避險)'].fillna(0)
```

```python
data=data.drop(['自營商買進股數(自行買賣)','自營商買進股數(避險)','自營商賣出股數(自行買賣)','自營商賣出股數(避險)'],axis=1)
```

```python
data['自營商買進股數'].fillna(data['自營商買進tmpsum'])
data['自營商賣出股數'].fillna(data['自營商賣出tmpsum'])
data
```
